# Adjusting the Plots

```python
plt.plot(x,np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.ylim(-1.5,1.5)
plt.xlim(-1,11)
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.grid()
plt.show()

```

```python
plt.plot(x, np.sin(x))
plt.grid()
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.show()
```

```python
plt.plot(x, np.sin(x),'-g',label='sin(x)')
plt.plot(x, np.cos(x),':b', label='cos(x)')
plt.title("A Sine and Cos Curve")
plt.legend()

plt.show()
```

# Histogram

```python
data=np.random.randn(10000)
plt.hist(data)#bins default is 10
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=30)#bins default is 10
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,density=True)#bins default is 10
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,density=True,alpha=0.1)#bins default is 10
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,density=True,alpha=0.5)#bins default is 10
plt.show()
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,density=True,alpha=0.9)#bins default is 10
plt.show()
#alpha is the Transperency value
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,density=True,alpha=0.5,histtype='stepfilled',color='steelblue')#bins default is 10
plt.show()
#alpha is the Transperency value
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,density=True,alpha=0.5,histtype='stepfilled',color='steelblue',edgecolor='b')#bins default is 10
plt.show()
#alpha is the Transperency value
```
